```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('dark_background')
```

```python
classes=['Physics', 'Chemistry', 'Maths','Science', 'SST']
marks=[89,45,78,23,90]
```

```python
plt.pie(marks,labels=classes)
plt.show()
```

```python
colors=['red','blue','green','#9803fc','#03c2fc']

plt.pie(marks,labels=classes,colors=colors)
plt.show()
```

```python
colors=['red','blue','green','#9803fc','#03c2fc']

plt.pie(marks,labels=classes,colors=colors, autopct='%0.2f%%')
plt.show()
```

```python
explode_values=[0.1,0.2,0,0,0]

plt.pie(marks,labels=classes,colors=colors, autopct='%0.1f%%', explode=explode_values)
plt.show()
```

```python
explode_values=[0.1,0.2,0,0,0]

plt.pie(marks,labels=classes,colors=colors, autopct='%0.1f%%', explode=explode_values, shadow=True)
plt.show()
```

```python
explode_values=[0.1,0.2,0,0,0]

plt.pie(marks,labels=classes,colors=colors, autopct='%0.1f%%', explode=explode_values, shadow=True, radius=1.6)
plt.show()
```

```python
textprops={'fontsize':14, 'color':'k'}

explode_values=[0.1,0.2,0,0,0]

plt.pie(marks,labels=classes,colors=colors, autopct='%0.1f%%', explode=explode_values, shadow=True, textprops=textprops)
plt.show()
```

```python
textprops={'fontsize':14, 'color':'k'}
wedgeprops={'linewidth':3,'linestyle':'--','edgecolor':'white'}

explode_values=[0.1,0.2,0,0,0]

plt.pie(marks,labels=classes,colors=colors, autopct='%0.1f%%', explode=explode_values, shadow=True, textprops=textprops, wedgeprops=wedgeprops)
plt.show()
```

```python
explode_values=[0.1,0.2,0,0,0]

plt.pie(marks,labels=classes,colors=colors, autopct='%0.1f%%', explode=explode_values, shadow=True, radius=1.6)
plt.title('Subjects and Average scores')
plt.legend()
plt.show()
```
