```python
import seaborn as sns
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic_df=sns.load_dataset('titanic')
titanic_df.head()
```

```python
sns.barplot(x='class',y='fare',data=titanic_df)
plt.show()
```

```python
sns.barplot(x='class',y='fare',data=titanic_df, hue='sex')
```

```python
sns.barplot(x='class',y='fare',data=titanic_df,
            hue='sex'
            ,palette='inferno')
```

```python
sns.barplot(x='fare',y='class',data=titanic_df,
            hue='sex'
            ,palette='inferno'
            ,orient='h') #should be numeric x=h
```

```python
sns.barplot(x='class',y='fare',data=titanic_df,
            hue='sex'
            ,palette='inferno',
            estimator=np.max)
```

```python
#confidence interval (uncertainty)
sns.barplot(x='class',y='fare',data=titanic_df,
            hue='sex'
            ,palette='inferno',
            errorbar=('ci', 100),
            err_kws={'color': 'red','linewidth': 10})
```

```python
sns.barplot(x='class',y='fare',data=titanic_df,
            hue='sex'
            ,palette='inferno',
            errorbar=('ci', 100),
            err_kws={'color': 'red','linewidth': 10},
            saturation=0.9)
```
