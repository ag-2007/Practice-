```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('dark_background')
```

```python
subjects=['Maths','English','Science','Social Studies','Computer']
marks=[89, 90, 45, 78, 99]
```

```python
plt.bar(subjects, marks, color='blue')
plt.show()
```

```python
plt.bar(subjects, marks, color='green')
plt.show()
```

```python
colors=['red','blue','green','orange','purple']
plt.bar(subjects, marks, color=colors)
plt.show()
```

```python
plt.bar(subjects,marks, color=colors, width=0.6)
plt.show()
```

```python
plt.bar(subjects,marks, color=colors, width=0.6, edgecolor='white')
plt.show()
```

```python
plt.bar(subjects,marks, color=colors, width=0.6, edgecolor='white', linewidth=4)
plt.show()
```

```python
plt.bar(subjects,marks, color=colors, width=0.6, edgecolor='white', linewidth=4, linestyle='--')
plt.show()
```

```python
plt. barh(subjects, marks, color=colors)
plt.show()
```

```python
subjects=['Maths','English','Science','Social Studies','Computer']
marks1=[89, 90, 45, 78, 99]
marks2=[78,56,34,90,12]
```

```python
plt.figure(figsize=(8,8))

plt.bar(subjects,marks1)
plt.bar(subjects, marks2)

plt.xlabel('Subject')
plt.ylabel('Marks')

plt.show()
```

```python
subjects_len=np.arange(len(subjects))
width=0.4
```

```python
plt.figure(figsize=(8,8))

plt.bar(subjects_len,marks1, width=width)
plt.bar(subjects_len+width, marks2, width=width)

plt.xlabel('Subject')
plt.ylabel('Marks')

plt.show()
```

```python
plt.figure(figsize=(8,8))

plt.bar(subjects_len,marks1, width=width)
plt.bar(subjects_len+width, marks2, width=width, alpha=0.5)

plt.xlabel('Subject')
plt.ylabel('Marks')

plt.show()
```
