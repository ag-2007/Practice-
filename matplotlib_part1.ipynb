```python
temperature_pune=[25,34,21,45,28,6,43,18,7,2]
humidity_pune=[28,25,29,20,26,50,19,29,52,55]

temperature_banglore=[34,35,36,37,28,27,26,25,31,20]
humidity_banglore=[40,38,36,35,42,44,41,40,34,45]
```

```python
plt.figure(figsize=(8,8))
plt.scatter(temperature_pune, humidity_pune, color='red')
plt.show()
```

```python
plt.figure(figsize=(8,18))
plt.xticks(np.arange(0,60,5))
plt.yticks(np.arange(10,60,5))

plt.scatter(temperature_pune, humidity_pune, color='red',marker='v')

plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.show()
```

```python
plt.figure(figsize=(8,18))
plt.xticks(np.arange(0,60,5))
plt.yticks(np.arange(10,60,5))

plt.scatter(temperature_pune, humidity_pune, color='red',marker='v')
plt.scatter(temperature_banglore, humidity_banglore, color='blue',marker='v')

plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.show()
```
